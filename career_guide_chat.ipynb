```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import gradio as gr

tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-base")
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base")
```

```python
def get_career_advice(user_input):
    prompt = f"Suggest 3 ideal careers based on this interest: {user_input}"
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(**inputs, max_new_tokens=100)
    reply = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return reply
```

```python
with gr.Blocks(theme=gr.themes.Soft()) as app:
    gr.Markdown("## 🎓 Career Compass — *Your Personal AI Career Guide*")
    gr.Markdown("Tell me what you love, and I’ll help you discover careers that match your passion 💡")

    chatbot = gr.Chatbot(label="Career Advisor Chat", type="messages")

    msg = gr.Textbox(label="Type your interest or passion here ✍️", placeholder="e.g., I love technology and building apps")
    state = gr.State([])

    def user_interact(message, history):
        history = history or []
        reply = get_career_advice(message)
        history.append({"role": "user", "content": message})
        history.append({"role": "assistant", "content": reply})
        return history, history

    msg.submit(user_interact, [msg, state], [chatbot, state])
    gr.Button("🔁 Reset Chat").click(lambda: ([], []), outputs=[chatbot, state])

    gr.Markdown("#### ✨ Or start by clicking a topic you like:")
    with gr.Row():
        gr.Button("💻 Technology").click(lambda: "I like technology and building things", outputs=msg)
        gr.Button("🎨 Art & Design").click(lambda: "I love drawing and creativity", outputs=msg)
        gr.Button("🧪 Science").click(lambda: "I'm interested in biology and experiments", outputs=msg)
        gr.Button("💼 Business").click(lambda: "I enjoy organizing and leading people", outputs=msg)
app.launch(share=True, debug=True)
```
